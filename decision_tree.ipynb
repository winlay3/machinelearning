```python
import numpy as np 
import pandas as pd 
#import graphviz
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
#from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
```

```python
dataset = load_iris(as_frame=True)
df = pd.DataFrame(data = dataset.data)
df.head()
```

```python
# adding the target and target names to dataframe
target_zip= dict(zip(set(dataset.target), dataset.target_names))
df["target"] = dataset.target
df["target_names"] = df["target"].map(target_zip)

print(df.shape)
df.head()
```

```python
# Seperating to X and Y 
X = df.iloc[:, :4]
y = df.iloc[:, -1]
```

```python
# splitting training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, shuffle=True, random_state=24)

print("X_train shape: ",X_train.shape)
print("y_train shape: ",y_train.shape)

print("X_test shape: ",X_test.shape)
print("y_test shape: ",y_test.shape)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
predictions = dt.predict(X_test)
print(predictions)
```

```python
N = 10
predicted_label = predictions[N]
predicted_label
y_test = np.array(y_test)
print("True label : ",y_test[N])
print("Predicted label : ",predicted_label)
```

```python
from sklearn.metrics import accuracy_score
print("My Implementation:\nACCURACY: {}".format(accuracy_score(y_test, predictions)))
```

## Decision Tree Framework with Image

```python
import numpy as np 
import pandas as pd 
import os
```

```python
root_path = r"C:\Users\User\Downloads\fruits\fruits"

kiwi = []
kiwi_path = root_path+"/Kiwi"
for i in os.listdir(kiwi_path):
    kiwi.append(kiwi_path+"/"+i)

pear = []
pear_path = root_path+"/Pear"
for i in os.listdir(pear_path):
    pear.append(pear_path+"/"+i)

watermelon = []
watermelon_path = root_path+"/Watermelon"
for i in os.listdir(watermelon_path):
    watermelon.append(watermelon_path+"/"+i)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = kiwi[0]
sample_img = cv2.imread(path)

print(sample_img.shape)

reshaped_arr = sample_img.reshape(-1)
print(reshaped_arr.shape)
print(len(reshaped_arr))
```

```python
def load_img(img_path_list,label):
    x, y = [],[]
    for cur_path in img_path_list:
        cur_img = cv2.imread(path)
        if cur_img is not None:
            cur_img = cv2.resize(cur_img,(224,224))
            reshaped_arr = sample_img.reshape(-1)
            x.append(reshaped_arr)
            y.append(label)
    return x,y
```

```python
x1, y1 = load_img(kiwi,1)
x2, y2 = load_img(pear,2)
x3, y3 = load_img(watermelon,3)


x,y = [],[]
for cur_x,cur_y  in zip(x1,y1):
    x.append(cur_x)
    y.append(cur_y)

for cur_x,cur_y  in zip(x2,y2):
    x.append(cur_x)
    y.append(cur_y)

for cur_x,cur_y  in zip(x3,y3):
    x.append(cur_x)
    y.append(cur_y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# splitting training and test
X_train, X_test, y_train, y_test = train_test_split(x,y, train_size=0.8, shuffle=True, random_state=240)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
predictions = dt.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
print("My Implementation:\nACCURACY: {}".format(accuracy_score(y_test, predictions)))
```
